```python
def sigmoid_normal(x):
    temp = math.exp(x)
    sig = temp/(1 + temp)
    
    return sig
```

```python
def sigmoid_approximate(x):
    
    temp1 = ((x/4) + 1)**2
    temp2 = ((x/4) - 1)**2
    if x <= -4:
        sig = 0
    elif x > -4 and x < 0:
        sig = temp1/2
    elif x >= 0 and x < 4:
        sig = 1 - (temp2/2)
    else:
        sig = 1
    
    return sig
```

```python
normal = list()
j = np.linspace(-8, 8, num=101)
for i in range(0,101):
    normal.append(sigmoid_normal(j[i]))
normal = np.array(normal)
x = np.linspace(-20, 20, num=101)
plt.plot(x,normal)
```

```python
approximate = list()
for i in range(0,101):
    approximate.append(sigmoid_approximate(j[i]))
approximate = np.array(approximate)
plt.plot(x,approximate)
```

```python
f = plt.figure()
f.set_figwidth(13)
f.set_figheight(13)

plt.plot(x, normal, '--b', c = 'blue', label ='Sigmoid')
plt.plot(x, approximate, c = 'red', label ='Approximation')

plt.legend()
plt.show()
```

```python
def Twoscomp(str):
    n = len(str)

    i = n - 1
    while(i >= 0):
        if (str[i] == '1'):
            break
 
        i -= 1

    if (i == -1):
        return '1'+str

    k = i - 1
    while(k >= 0):
         
        if (str[k] == '1'):
            str = list(str)
            str[k] = '0'
            str = ''.join(str)
        else:
            str = list(str)
            str[k] = '1'
            str = ''.join(str)
 
        k -= 1

    return str
```
